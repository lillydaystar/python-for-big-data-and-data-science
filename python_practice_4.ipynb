6. (7б) Визначте функцію, яка приймає iterable (ітерабельний обʼєкт, напр. список), що складається з 7ми рядкових значень - назв погодних умов (напр. "sunny") за кожен день у тижні. Повертає функція словник, де ключами є рядки з погодними умовами, а значеннями - кількість разів, скільки дана погода була протягом тижня (напр. {'sunny': 1, ...}). Викличте цю функцію та виведіть результат її роботи на екран.

```python
def dictionary(iterable_obj):
  weather = {}
  for elem in iterable_obj:
    if elem in weather:
      weather[elem] += 1
    else:
      weather[elem] = 1
  return weather

week = ['sunny', 'cloudy', 'rainy', 'sunny', 'windy', 'cloudy', 'sunny']
w_dict = dictionary(week)
print(w_dict)
```

7. (15б) Напишіть програму, яка б шукала необхідне число у впорядкованому списку цілочислених значень, де
число має вводитись з консолі.
власне алгоритм пошуку винесений в окрему функцію, яка повертає булеве значення (знайдено число - True, не знайдено - False).
в залежності від результату функції, виведіть на екран повідомлення про успішний або ж неуспішний пошук.
Зауваження: використовувати оператор in заборонено.

```python
def find_num(num, search_list):
  left = 0
  right = len(search_list) - 1

  while left <= right:
    mid = (left + right) // 2
    if search_list[mid] == num:
      return True
    elif search_list[mid] < num:
      left = mid + 1
    else:
      right = mid - 1
  return False

search_list = [-1, 3, 5, 8, 13, 23, 26, 43, 52, 54, 56, 78, 91, 99]
num = int(input("Введіть число для пошуку: "))
res = find_num(num, search_list)
if res:
  print("Ваше число присутнє в списку :)")
else:
  print("Вашого числа немає в списку :(")
```

# III. Decorators.

11. (11б) Визначте декоратор, який би виводив у консоль повідомлення про початок роботи і назву функції (напр. my_function._ _ name _ _), а після відпрацювання функції - повідомлення про закінчення роботи і назву функції, що задекорована.

```python
import time

def start_end_decorator(my_function):
    def wrapper(*args, **kwargs):
        print(f"Початок роботи функції {my_function.__name__}")
        begin = time.time()
        result = my_function(*args, **kwargs)
        end = time.time()
        print(f"Закінчення роботи функції {my_function.__name__} за час:", end - begin, "с")
        return result
    return wrapper

@start_end_decorator
def long_function():
  x = 0
  print("doing smth")
  while x < 100000000:
    x += 1

long_function()
```

12. (4б) Задекоруйте двома різними способами функції з завдання 4 та 5 декоратором із завдання 11. (**УВАГА**: не декоруйте функції двічі!).

```python
# 1й спосіб (потрібно перевизначити функцію)
@start_end_decorator
def divide_none(arg1, arg2):
  if arg2 == 0:
    return None
  return arg1 // arg2

print(divide_none(10,0))

# 2й спосіб без перевизначення
diff = start_end_decorator(diff)

print(diff(10))
```

15. (15б) Визначте функцію, яка рекурсивно реалізовує пошук числа у послідовності Фібоначчі за її індексом (нумерація починається з нуля). Наприклад, потрібно знайти перше число у послідовності, тоді функція прийматиме 0 у якості аргумента і повертатиме 1, якщо ж знаходити третє число (індекс 2) послідовності, то функція повертатиме 2, якщо ж знаходити десяте число (індекс 9), то функція повертатиме 55.

```python
def fib_rec(n):
  if n == 0:
    return 1
  elif n == 1:
    return 1

  return fib_rec(n-1) + fib_rec(n-2)

print(fib_rec(9))
```
